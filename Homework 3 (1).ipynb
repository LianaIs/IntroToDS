```python
# modules we'll use
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# read in 2 datasets provided
nfl_data = pd.read_csv('C:/Users/ASUS/Downloads/NFL Play by Play 2009-2016 (v3) (1).csv')
pd.set_option('display.max_columns', None) # What does this option do? Answer below in a comment

# is used for expanding the output display to see more columns of dataframe
```

```python
# look at a few rows of the data
nfl_data.head()
```

```python
# get the number of missing data points per column
missing_values_count = nfl_data.isna().sum()
print(missing_values_count[0:10])
```

```python
# how many total missing values do we have?
total_missing = nfl_data.isna().sum().sum()

# percent of data that is missing
num_total_cells =  nfl_data.count().sum()
percent = total_missing * 100  / num_total_cells
percent
```

```python
# remove all columns with at least one missing value
columns_with_na_dropped = nfl_data.dropna(axis='columns')
columns_with_na_dropped.head()
```

```python
# just how much data did we lose? 
print("Columns in original dataset: %d \n" % len(nfl_data.columns))
print("Columns with na's dropped: %d" % len(columns_with_na_dropped.columns))
```

```python
# replace all NA's with 0
filled = nfl_data.fillna(0)
filled.head()
```

```python
# replace all NA's the value that comes directly after it in the same column, 
# then replace all the reamining na's with 0
bfilled = nfl_data.fillna(method='bfill').fillna(0)
bfilled.head()
```

```python
# Create indicator column for missing values in data nlf_data for the column 'down'
nfl_data['indicator'] = np.select(nfl_data.down.isnull(),nfl_data.down.isnull())
```

```python
# print columns that have more than 50 missing values in data nlf_data
column_with_nan = nfl_data.columns[nfl_data.isna().any()]
for column in column_with_nan:
    if nfl_data[column].isna().sum() > 50:
        print(column)
```
